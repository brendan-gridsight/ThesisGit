```python
from __future__ import print_function
import os
import sys
import numpy as np
import pandas as pd
from pandas import DataFrame
from __main__ import *
from scipy.stats import norm
import matplotlib
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns
```

```python
sns.set(font_scale=1.35, rc={'figure.figsize':(12,5)})
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
matplotlib.rcParams['font.family'] = 'Abyssinica SIL'
```

```python
df = pd.read_csv("price.csv")

fig, ax = plt.subplots()
ax.set_xlim(0, 47)

first = plt.plot(df.time24, df.TOU, linewidth=2, label='$C_{t}$ - TOU Tariff')
second = plt.plot(df.time24, df.FIT, linewidth=2, label='$C_{f}$ - Feed-in Tariff', color='red', linestyle='--')
third = plt.plot(df.time24, df.batt, linewidth=2, label='$C_{b}$ - Battery Cost',linestyle='-.')

# plt.rc('font', family='Times New Roman')
plt.xticks(np.arange(0, 48, step=6), fontsize=14)
plt.yticks(np.arange(0, 0.4, step=0.05), fontsize=14)

# for tick in ax.get_xticklabels():
#     tick.set_fontname("Times New Roman")
# for tick in ax.get_yticklabels():
#     tick.set_fontname("Times New Roman")

#plt.text(-4.5,65,"Set \n Point",family='Times New Roman',horizontalalignment='center', fontsize=12) #place neg sigma symbol on graph
plt.xlabel('\nTime', fontsize=16, fontweight='bold')
plt.ylabel('Cost ($/kWh)', fontsize=16,fontweight='bold')
#plt.title("Net Power 442 Customers 25-01-2019", family='Times New Roman')
#plt.legend()
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.00), shadow=True, ncol=3)
plt.grid(True)
plt.savefig('tariffs.pdf', bbox_inches='tight')
plt.show()
```

```python
df = pd.read_csv("price.csv")

fig, ax = plt.subplots(figsize=(13, 5))
ax.set_xlim(0, 47)

first = plt.plot(df.time24, df.TOU, linewidth=2, label='$C_{t}$ - TOU ')
second = plt.plot(df.time24, df.FIT, linewidth=2, label='$C_{f}$ - Feed-in', color='red', linestyle='--')
third = plt.plot(df.time24, np.ones(len(df.time24))/100, linewidth=2, label='$C_i^{b}$ - Local ES',linestyle='-.')
third = plt.plot(df.time24, np.ones(len(df.time24))/50, linewidth=2, label='$C_i^{g}$ - Grid ES',linestyle='dotted')

# plt.rc('font', family='Times New Roman')
plt.xticks(np.arange(0, 48, step=6), fontsize=14)
plt.yticks(np.arange(0, 0.4, step=0.05), fontsize=14)



plt.xlabel('\nTime', fontsize=16, fontweight='bold')
plt.ylabel('Cost ($/kWh)', fontsize=16,fontweight='bold')
plt.ylim(0,0.38)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.00), shadow=True, ncol=4)
plt.grid(True)
plt.savefig('tariffs_drh.pdf', bbox_inches='tight')
plt.show()
```
