```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import matplotlib
import numpy as np
import seaborn as sns
import datetime
#from datetime import datetime
import time
```

```python
ps_df = pd.read_csv('PeakShaving_Comparison.csv', index_col=0)
ps_df.index = pd.to_datetime(ps_df.index, format='%Y-%m-%d %H:%M:%S')
```

# Before Control

```python
sns.set(font_scale=1.35, rc={'figure.figsize':(12,5)})
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
matplotlib.rcParams['font.family'] = 'Abyssinica SIL'

fig, ax = plt.subplots()


before = ax.plot(ps_df.index, ps_df['Before'], color='blue', linestyle='--', linewidth=1.5, label='Before')
# after = ax.plot(ps_df.index, ps_df['DRM'], linewidth=3, color='orange',linestyle=':', label='DRM')
# cmpc = ax.plot(ps_df.index, ps_df['CMPC'], linewidth=3, label='CMPC', color='g',linestyle='--')
# droop = ax.plot(ps_df.index, ps_df['DMPC'], linewidth=3, label='DMPC')
sp = ax.plot(ps_df.index, np.ones(len(ps_df.index))*80, 'k',linestyle=':', linewidth=1.5, label='Set Point')

ax.set_ylabel('Power (kW)')
ax.set_xlabel('Time')
ax.set_ylim(-20,120)
ax.set_xlim(ps_df.index[0],ps_df.index[-1])

ax.legend(['Before', 'Set Point'], loc="upper left", ncol=2)

fig.tight_layout()

plt.savefig('load_before.pdf', bbox_inches='tight')

plt.show()
```

```python
fig, ax = plt.subplots()


before = ax.plot(ps_df.index, ps_df['Before'], color='blue', linestyle='--', linewidth=1.5, label='Before')
after = ax.plot(ps_df.index, ps_df['DRM'], linewidth=1.5, color='green',linestyle='-.', label='DRM')
sp = ax.plot(ps_df.index, np.ones(len(ps_df.index))*80, 'k',linestyle=':', linewidth=1.5, label='Set Point')

ax.set_ylabel('Power (kW)')
ax.set_xlabel('Time')
ax.set_ylim(-20,120)
ax.set_xlim(ps_df.index[0],ps_df.index[-1])

ax.legend(['Before', 'DRM', 'Set Point'], loc="upper left", ncol=3)

fig.tight_layout()

plt.savefig('load_drm_shave.pdf', bbox_inches='tight')

plt.show()
```
