```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib
```

```python
m_cool, b_cool = np.polyfit(x=[20,25,30,35,40,45], y=[-10.24, -9.38, -8.41, -7.44, -5.16, -3.79], deg=1)
```

```python
m_cop, b_cop = np.polyfit(x=[20,25,30,35,40,45], y=[3.59, 3.26, 2.85, 2.44, 2.15, 1.88], deg=1)
```

```python
t_amb = [20,25,30,35,40,45]
cool = [-(m_cool*t + b_cool) for t in t_amb]
cop = [(m_cop*t + b_cop) for t in t_amb]
cool
```

```python
sns.set(font_scale=1.35, rc={'figure.figsize':(12,5)})
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
matplotlib.rcParams['font.family'] = 'Abyssinica SIL'


#Plot Results
fig, ax1 = plt.subplots()
ax1.set_xlabel('Ambient Temperature ($^{\circ} C$)')
ax1.set_ylabel('Maximum Cooling Capacity (kW)')
ax1.set_ylim(0,12)
#ax1.plot(df_results.index, df_results['phi_h_heat'], color='red')
ax1.plot(t_amb, cool, lw=2)
ax1.legend(['Cooling Capacity'], loc="lower left")

ax2 = ax1.twinx()
ax2.set_ylabel('COP')
ax2.set_xlim(20,45)
ax2.set_ylim(0,5)
ax2.plot(t_amb, cop, color='red', lw=2)
ax2.legend(['COP'], loc="lower right")

fig.tight_layout()
fig.set_figheight(5)
fig.set_figwidth(12)# otherwise the right y-label is slightly clipped
ax2.grid()

plt.savefig('hp_coolcap_cop.pdf', bbox_inches='tight')

plt.show()
```
